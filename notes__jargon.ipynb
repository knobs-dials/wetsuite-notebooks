## Purpose of this notebook

This is a list of legal and technical jargon (and abbreviations I keep seeing), only really added when understanding them was necessary for this task.

In particular terms that 
- might be new to law students, 
- might be new to programmers, 
- might be new to to both, 
- might have have different meanings between these fields (e.g. annotation)

DUE WARNING: this was written by a programmer, so particularly legal terms it may be oversimplified, incorrect, or both.



## A

**Akomo Ntoso**  - XML document format aimed at executive, legislative and judiciary documents, with a controlled vocabulary
- [wikipedia link](https://en.wikipedia.org/wiki/Akoma_Ntoso) [akomantoso.org](http://www.akomantoso.org/)

**AKN** - often refers to the [Akoma Ntoso Naming Convention (AKN NC)](http://docs.oasis-open.org/legaldocml/akn-nc/v1.0/akn-nc-v1.0.html)
- which are URN-like construction that look like paths, like `/akn/nl/act/gm0202/2023/Regeling3d27a1736497475a8534a65837a94519`
- where e.g. the second item is the country, the third is usually one of a handful of doctypes, and there are a bunch of optional things
- See also JOIN
- https://koop.gitlab.io/STOP/standaard/1.3.0/naamgevingsconventie.html

**AKN resolver** (arguably more fully Akoma Ntoso IRI Resolver) - a service that resolves AKBs (and variants, see JOIN) do documents within a system
- a number of resolvers exist, and may be specific to a specific doctype (e.g. /akn/nl/bill) and/or the system/dataset they sit next to, so you may need to query multiple to get an answer you consider complete, and you may need to find some sort of catalogue of options to find out where (see e.g. [this](https://repository.overheid.nl/frbr/cga/resolver/performers/1/json/servers.json))
- https://akresolver.cs.unibo.it/ https://www.overheid.nl/resolver/introductie

**Avv** (algemeen verbindende voorschriften) - roughly speaking, a rule that applies to everyone, and is binding, usually based on a specific law (VERIFY)
- (more practical contrast somewhere below)

Algemene wetten - general laws laws that give broad rules for a larger area. Consider e.g. [those mentioned here (Aanwijzingen voor de regelgeving - Aanwijzing 2.46. Afwijking van algemene wetten)](https://wetten.overheid.nl/BWBR0005730/2022-04-01/#Hoofdstuk2_Paragraaf2.8_Artikel2.46)

**AMvB** (Algemene Maatregel van Bestuur)
- a few are independent (zelfstandige AMvB), but most are delegated legislation (gedelegeerde wetgeving), meaning they come from a specific law that says "that details have to be worked out separately"
- often ''relatively'' practical implementation details, like the method of distributing finances, how executive groups should be formed, etc.
- implicitly makes it easier to create the law itself, as well as to change these implementation details, as AMvBs are easier to add and change implementation details of <!--, e.g. to change drugs to the list the Opiumnwet applies to[https://zoek.officielebekendmakingen.nl/stb-2017-206.html]) than laws -->, e.g. often does not require [Staten-Generaal](info_jargon.ipynb#S), though should still involve [Raad van State](info_jargon.ipynb#R) (VERIFY)
- published in Staatsblad

**Annotation**
- In the legal world, annotation is a pragmatic human thing, a note to explain or illustrate meaning or help interpret a text, of moderate sized chunks, sometimes sometimes of the whole
- In the computer/data world, annotation is a mechanical thing, often demarking specific things to extract, often for training data to recognize similar elsewhere (e.g. 'this is a reference to a law'), often small indivisible things so that little to no fuzziness or interpretation is necessary for machine learning that uses it

**Beslissing** - decision, in a neutral sense and not necessarily relating to legalities at all.  Even in legal context, beslissing may refer to nothing more than deciding to get coffee.  (VERIFY)

**Besluit** - a beslissing(decision) that has some sort of legal implication (rechtsgevolg), that _changes_ something.
- Often a yes-or-no sort of thing.  (VERIFY)
- pretty central in bestuursrecht (e.g. has a definition at [Artikel 1:3 Algemene wet bestuursrecht](https://wetten.overheid.nl/jci1.3:c:BWBR0005537&hoofdstuk=1&titeldeel=1.1&artikel=1:3&z=2018-01-01&g=2018-01-01))
  - may be mentioned as a besluit van algemene strekking (bas), to contrast it with one that is ''not'', i.e. a beschikking

**Beschikking** - besluit that applies to a specific case (rather than something that might apply in general)
- e.g. handling a specific person's case
- or handling a specific permit (in bestuursrecht, administrative law)
- or, in civiel recht (civil law), the final say in a procedure that starts with a verzoekschrift<!--
(petition?)  (instead of starting with a  dagvaarding(summons), which leads to a vonnis(judgment)).
-->

**Consolidatie**  
- For context: 
  - a lot of legal text will be modified once or more. 
  - for laws this often comes in the form of _another_ law that says it augments the earlier version with a few details. The text of this agumenting law will be only those changes.
  - The _combination_ of these separate-but-together laws is what is really in effect, but hard to oversee, particularly for non-experts.
- Sometimes it us useful to present all these components as a set
- More usually, people care about what they have to adhere to today, so we prefer to see consolidations: a single single text with all updates included
  - Consolidations are considered informative, not legally binding, _roughly_ because it is considered redundant with the basic laws
  - but they _should_ be equivalent, so you can basically treat them as if they are
- When we look at resources that serve us text, it is useful to know whether it publishes the changes and/or the consolidated version
